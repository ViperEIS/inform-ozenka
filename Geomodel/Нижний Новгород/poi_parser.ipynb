##### Schools

```python
addresses = list()
titles = list()
```

```python
urls = ['https://nnmama.ru/guide/education/nnschools/?page=1',
       'https://nnmama.ru/guide/education/nnschools/?page=2',
       'https://nnmama.ru/guide/education/nnschools/?page=3',
       'https://nnmama.ru/guide/education/nnschools/?page=4',
       'https://nnmama.ru/guide/education/nnschools/?page=5',
       'https://nnmama.ru/guide/education/nnschools/?page=6',
       'https://nnmama.ru/guide/education/nnschools/?page=7',
       'https://nnmama.ru/guide/education/nnschools/?page=8',
       'https://nnmama.ru/guide/education/nnschools/?page=9',
       'https://nnmama.ru/guide/education/nnschools/?page=10']
```

```python
for el in urls:
    page = requests.get(el)
    soup = BeautifulSoup(page.text, "html.parser")
    for i in soup.find_all('div', class_= 'organization__field')[::2]:
        if len(i) == 6:
            addresses.append(i.get_text().split('\n')[2:4])
        else:
            addresses.append(i.get_text().split('\n')[2])
    titles += [el.get_text() for el in soup.find_all('a', class_='organization__name')]
```

```python
df_schools = pd.DataFrame(np.array([addresses, titles]).transpose(), columns=['Address', 'Title'])
df_schools.to_csv('Schools.csv', index=False)
```

```python
df_schools = pd.read_csv('Schools.csv')
gf.convert_df_address_to_coord(df_schools, 'Нижний Новгород', 'School')
```

```python
addresses = list()
titles = list()
```

```python
urls = ['https://nnmama.ru/guide/education/sady/?page={}'.format(ind) for ind in range(1, 19)]
```

```python
df_kindergartens = pd.DataFrame(np.array([addresses, titles]).transpose(), columns=['Address', 'Title'])
df_kindergartens.to_csv('Kindergartens.csv', index=False)
```

##### Polyclinics

```python
addresses = list()
titles = list()
```

```python
urls = ['https://prodoctorov.ru/nnovgorod/top/poliklinika/?page=1',
       'https://prodoctorov.ru/nnovgorod/top/poliklinika/?page=2',
       'https://prodoctorov.ru/nnovgorod/top/poliklinika/?page=3']
```

```python
for el in urls:
    page = requests.get(el)
    soup = BeautifulSoup(page.text, "html.parser")
    addresses += [el.get_text() for el in soup.find_all('div', class_='b-card__address')]
    titles += [el.get_text() for el in soup.find_all('a', class_='b-card__name-link')]
```

```python
df_polyclinics = pd.DataFrame(np.array([addresses, titles]).transpose(), columns=['Address', 'Title'])
df_polyclinics['Title'] = df_polyclinics['Title'].apply(str.strip)
df_polyclinics.to_csv('Polyclinics.csv', index=False)
```

```python
df_polyclinics = pd.read_csv('Polyclinics.csv')
gf.convert_df_address_to_coord(df_polyclinics, 'Нижний Новгород', 'Polyclinic')
```
