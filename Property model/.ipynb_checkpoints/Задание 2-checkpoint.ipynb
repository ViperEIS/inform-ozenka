### Предобработка
#### (предоброботку можно было бы сделать через пайплайн и фанкшен трансформер, но я решил обойтись без них)

##### Обработка пропущенные значения

```python
data.isnull().sum()
```

```python
def perc_nans():
    print('Доля пропусков \'Основная ценовая зона\':', np.around(data['Основная ценовая зона'].isnull().sum() / data.shape[0], 3))
    print('Доля пропусков \'Станция метро\':', np.around(data['Станция метро'].isnull().sum() / data.shape[0], 3))
    print('Доля пропусков \'Широта\':', np.around(data['Широта'].isnull().sum() / data.shape[0], 3))
    print('Доля пропусков \'Долгота\':', np.around(data['Долгота'].isnull().sum() / data.shape[0], 3))
perc_nans()
```

```python
data['Основная ценовая зона'].fillna(data['Основная ценовая зона'].value_counts().max(),inplace=True)
data['Станция метро'].fillna(data['Станция метро'].value_counts().max(),inplace=True)
data['Широта'].fillna(data['Широта'].mean(),inplace=True)
data['Долгота'].fillna(data['Долгота'].mean(),inplace=True)
```

```python
perc_nans()
```

##### Кодировние категориальных признаков

```python
obj_data = data.select_dtypes(include=['object']).copy().drop(columns='Дата создания')
obj_data.head(20)
```

```python
obj_data['Основная ценовая зона'] = obj_data['Основная ценовая зона'].astype(str)
obj_data['Станция метро'] = obj_data['Станция метро'].astype(str)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for ser in obj_data:
    if obj_data[ser].dtype != 'int64':
        le.fit(obj_data[ser])
        obj_data[ser] = le.transform(obj_data[ser])
```

```python
obj_data.head(20)
```

##### Стандартизация числовых признаков

```python
data_y = data['Удельная цена, руб./кв.м']
num_data = data.select_dtypes(include=['float64']).copy().drop(columns='Удельная цена, руб./кв.м')
num_data.head(20)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
num_data = pd.DataFrame(StandardScaler().fit_transform(num_data), columns=num_data.columns)
```

```python
num_data.head(20)
```

```python
data = pd.concat([data['Дата создания'], np.round(num_data, 2), obj_data], axis=1)
```

```python
data.head(20)
```

```python
data['Дата создания'] = pd.to_datetime(data['Дата создания'])
```

##### Обработка значимости признаков (корреляционная матрица)

```python
colormap = plt.cm.Blues
```

```python
plt.figure(figsize=(12,12))
plt.title('Price correlation')
sns.heatmap(pd.concat([data, data_y], axis=1).corr(), linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor='white', annot=True)
```

```python
print(data['Город'].unique(), data['Сегмент'].unique())
data.drop(columns=['Город', 'Сегмент'], inplace=True)
```

### Построение моделей

```python
from sklearn.linear_model import Lasso, Ridge, LassoCV, RidgeCV
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor, RadiusNeighborsRegressor
from sklearn.model_selection import KFold, GridSearchCV, train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
data_x = data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data_x.drop(columns='Дата создания'), data_y, test_size=0.3)
k_fold = KFold(n_splits=4, shuffle=True, random_state=0)
```

##### Гребневая регрессия

```python
alphas = np.logspace(-2, 6, 200)
```

```python
ridge_cv = RidgeCV(alphas=alphas, cv=k_fold, scoring='r2').fit(X_train, y_train)
ridge_reg = Ridge(alpha=ridge_cv.alpha_).fit(X_train, y_train)
y_test_pred = ridge_reg.predict(X_test)
print('r2_score is:', r2_score(y_test_pred, y_test))
```

График, демонстрирующий, как изменяются веса регрессии в зависимости от изменения альфа

```python
coefs = []
for a in alphas:
    coefs.append(Ridge(a).fit(X_train, y_train).coef_)
    
plt.figure(figsize=(10, 8))
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.legend(X_train.columns)
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Alpha dependence of weights (Ridge)')
plt.axis('tight')
plt.show()
```

##### Лассо регрессия

```python
lasso_cv = LassoCV(alphas=alphas, cv=k_fold).fit(X_train, y_train)
lasso_reg = Lasso(alpha=lasso_cv.alpha_).fit(X_train, y_train)
y_test_pred = lasso_reg.predict(X_test)
print('r2_score is:', r2_score(y_test_pred, y_test))
```

```python
coefs = []
for a in alphas:
    coefs.append(Lasso(a).fit(X_train, y_train).coef_)
    
plt.figure(figsize=(10, 8))
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.legend(X_train.columns)
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Alpha dependence of weights (Lasso)')
plt.axis('tight')
plt.show()
```

######  Полиномиальная регрессия

```python
poly_model = make_pipeline(PolynomialFeatures(degree=2), Ridge())
grid_poly = GridSearchCV(poly_model, {'ridge__alpha': alphas}, scoring='r2', cv=k_fold).fit(X_train, y_train)
X_train_poly = PolynomialFeatures(degree=2).fit_transform(X_train)
poly_model = grid_poly.best_estimator_
y_test_pred = poly_model.predict(X_test)
print('r2_score is:', r2_score(y_test_pred, y_test))
```

График зависимости коэффициента детерминации от заданных альфа

```python
plt.plot(alphas, grid_poly.cv_results_['mean_test_score'], label='train', marker='.')
grid_poly.fit(X_test, y_test)
plt.plot(alphas, grid_poly.cv_results_['mean_test_score'], label='test', marker='.')
plt.title("R2 dependency params")
plt.xlabel('alphas')
plt.ylabel('R2')
plt.legend(loc="lower right")
plt.show()
```

##### KNN

```python
param_grid = {
            'n_neighbors': range(2, 6),
            'leaf_size': range(20, 35),
            'metric': ['minkowski', 'euclidean']
            }
grid_knn = GridSearchCV(KNeighborsRegressor(), param_grid, scoring='r2', cv=k_fold).fit(X_train, y_train)
knn_model = grid_knn.best_estimator_
y_test_pred = knn_model.predict(X_test)
print('r2_score is:', r2_score(y_test_pred, y_test))
```

Пример графика зависимости показателя коэффициента детерминации от заданных параметров (аналогично для остальных параметров)

```python
plt.plot(grid_knn.param_grid['n_neighbors'], grid_knn.cv_results_['mean_test_score'][::30], label='train', marker='.')
grid_knn.fit(X_test, y_test)
plt.plot(grid_knn.param_grid['n_neighbors'], grid_knn.cv_results_['mean_test_score'][::30], label='test', marker='.')
plt.title("R2 dependency params")
plt.xlabel('n_neighbors')
plt.ylabel('R2')
plt.legend(loc="lower right")
plt.show()
```

##### Деревья решений

```python
param_grid = {
            'max_depth': range(5, 15),
            'min_samples_split': range(2,5),
            'min_samples_leaf': range(1,3)
            }
grid_dt = GridSearchCV(DecisionTreeRegressor(), param_grid, scoring='r2', cv=4).fit(X_train, y_train)
dt_model = grid_dt.best_estimator_.fit(X_train, y_train)
y_test_pred = dt_model.predict(X_test)
print('r2_score is:', r2_score(y_test_pred, y_test))
```

Пример графика зависимости показателя коэффициента детерминации от заданных параметров (аналогично для остальных параметров)

```python
plt.plot(grid_dt.param_grid['max_depth'], grid_dt.cv_results_['mean_test_score'][::6], label='train', marker='.')
grid_dt.fit(X_test, y_test)
plt.plot(grid_dt.param_grid['max_depth'], grid_dt.cv_results_['mean_test_score'][::6], label='test', marker='.')
plt.title("R2 dependency params")
plt.xlabel('max_depth')
plt.ylabel('R2')
plt.legend(loc="lower right")
plt.show()
```
